```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from wordcloud import WordCloud
```

```python
df = pd.read_csv("data/resume_cleaned_dataset.csv")
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(
    y=df["category"],
    order=df["category"].value_counts().index
)
plt.title("Number of Resumes per Category")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df["word_count"], bins=20, edgecolor="black")
plt.title("Resume Length Distribution")
plt.xlabel("Word Count")
plt.ylabel("Count")
plt.show()
```

```python
avg_len = df.groupby("category")["word_count"].mean().sort_values()

plt.figure(figsize=(8,5))
avg_len.plot(kind="barh")
plt.title("Average Resume Length by Category")
plt.xlabel("Word Count")
plt.show()
```

```python
text = " ".join(df["clean_text"])

wc = WordCloud(
    width=1200,
    height=600,
    background_color="white"
).generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wc)
plt.axis("off")
plt.title("Resume Word Cloud")
plt.show()
```
